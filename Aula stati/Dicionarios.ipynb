Adicionar mais um item no dicionário compra

`````python
compra['id_102'] = {
    'quantidade':2,
    'produto':'camiseta',
    'preco':98.90,
    'embalagem':'-',
    'subtotal':98.90
}
`````

`````python
print(compra)
`````

`````python
produto_id_102 = compra['id_102']['produto']
print(produto_id_102)
`````

`````python
preco_id_101 = compra["id_101"]['preco']
preco_id_101
`````

`````python
embalagem_compra = compra['id_101']['embalagem']
print(embalagem_compra)
`````

`````python
#usando laço for para iterar todos os elementos do dicionário

for id_compra,dados_compra in compra.items():
    preco = dados_compra['preco']
    quantidade = dados_compra['quantidade']

    print(f'id_compra: {id_compra}')
    print(f'Preco: {dados_compra["preco"]}')
    print(f'Quantidade: {dados_compra["quantidade"]}')
    print(f'-'*20)
`````
